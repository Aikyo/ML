```python

# Least square compute residual
def ComputeResidual(x_data,y_data,w1,w2):
    tatolError = 0
    Length = len(x_data)
    for i in range(0,Length):
        tatolError += (y_data[i] - (w1*x_data[i] + w2))**2
    return tatolError / float(len(x_data)) / 2.0
ComputeResidual(x,y,0,0)
```
