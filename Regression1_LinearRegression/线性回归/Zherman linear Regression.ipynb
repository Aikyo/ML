```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv',delimiter=",",header = 0)

#x,y = data[["x","y"]]
X = data["x"]
Y = data["y"]

u = np.linspace(20,70,100)

plt.plot(u,1.4*u + 0.1,color='red')

plt.scatter(X,Y)
plt.show()
```

```python
def Least_Square(k,b,x_data,y_data):
    
    X = np.linspace(20,70,100)
    plt.plot(X,k*X + b,color='red')
    plt.scatter(x_data,y_data)
    
    totalError = 0
#   print(len(x_data))
    for i in range(0,len(x_data)):
        
        totalError += ( (k*x_data[i] + b) -y_data[i])**2
    return totalError/float(len(x_data))/2.0


print("---Least_square--------------")
MeanError = Least_Square(1.4,0,X,Y)
print('MeanError {0}'.format(MeanError))
print("-----------------")
```

```python


def Gradient_Descent(k,b,x_data,y_data,epochs,lr):
    
    m = float(len(x_data))

    for i in range(0,epochs):
        k_grad = 0
        b_grad = 0
        for j in range(0,len(x_data)):
            k_grad += (1/m) * ((k*x_data[j] + b) - y_data[j]) * x_data[j]
            b_grad += (1/m) * ( k*x_data[j] + b - y_data[j] )

        print(k_grad)
        print(b_grad)
        k = k - lr*k_grad
        b = b - lr*b_grad
    return k,b


print("------------------------")
        
M = Least_Square(0,0,X,Y)
print(M)

k,b = Gradient_Descent(0,0,X,Y,50,0.0001)
M2 = Least_Square(k,b,X,Y)
print('k = {0}  b = {1}  M2 = {2}'.format(k,b,M2))

    


```
