```python
x = np.linspace(-5,5,10)
x1 ,x2 = np.meshgrid(x,x)
z = np.sqrt(x1**2 + x2**2)

#
figure = plt.figure(figsize=(20,8))
ax = figure.add_subplot(231)
ax.plot(x,x**2)


ax = figure.add_subplot(231)
ax.plot(x1,x2)

ax = figure.add_subplot(233)
ax.imshow(np.sqrt(x1**2 + (x2 -5)**2),cmap = plt.cm.gray)

ax = figure.add_subplot(232)
ax.imshow(z,cmap = plt.cm.gray)

ax = figure.add_subplot(234)
ax.imshow(np.sqrt(x1**2 + (x2 -5)**2),cmap = plt.cm.cool)

ax = figure.add_subplot(235)
ax.imshow(z,cmap = plt.cm.hot)

plt.show()
```

```python
#zip work like a zip on two sequence
# zip() in conjunction with the * operator can be used to unzip a list:



plt.figure(figsize=(16,3))
for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
    plt.subplot(1, 5, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize = 20)
```
