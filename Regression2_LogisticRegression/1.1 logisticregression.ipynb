```python
# splitting data into trainning set and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size = 0.25,random_state=0)

```

```python
# step1 : from linear_model import logisticregression ,and make a instance 

from sklearn.linear_model import LogisticRegression 
logisticregression = LogisticRegression()

```

```python
# step2 : Trainning the model on the data , stored the information learned from the data.Model is learning the relationship between  digits and label
logisticregression.fit(x_train,y_train)
```

```python
# reshape(n,-1) 自动计算列
b = np.array(range(1,10)).reshape(1,-1)
a = np.arange(10).reshape(5,-1)
print(a)
print(b)
```

```python
# step 3 : prediction 
test1 = x_test[5].reshape(1,-1)
train1 = digits.data[4].reshape(1,-1)
print(train1)


logisticregression.predict(train1)
logisticregression.predict(x_test[0:12])
logisticregression.predict(x_test)
score = logisticregression.score(x_test,y_test)
print(score)
```
