```python
def Draw_Scatter(x_data,y_data,k,b):
    plt.scatter(x_data,y_data)
    u = np.linspace(20,70,100)
    plt.plot(u,k*u + b,color='red')

Draw_Scatter(X,Y,1,0.1)

    
```

```python
def Draw_Linear(k,b):
    u = np.linspace(20,70,100)
    plt.plot(u,k*u + b,color='red')
```

```python
def Least_Square(k,b,x_data,y_data):
    
    X = np.linspace(20,70,100)
    plt.plot(X,k*X + b,color='red')
    
    plt.scatter(x_data,y_data)
    
    totalError = 0
    #length = len(x_data)
    #print('数据集大小 : {0}'.format(length)
    
    for i in range(0,len(x_data)):
        totalError += ( (k*x_data[i] + b) -y_data[i])**2
    return totalError/float(len(x_data))/2.0

meanError = Least_Square(-7,1,X,Y)
print(meanError)
```

```python


print("---gradient--------------")

def Gradient_Descent(k,b,x_data,y_data,epochs,lr):
    
    #plt.scatter(x_data,y_data)
    m = float(len(x_data))
    for i in range(0,epochs):
        k_grad = 0
        b_grad = 0
        for j in range(0,len(x_data)):
            k_grad += (1/m) * ((k*x_data[j] + b) - y_data[j]) * x_data[j]
            b_grad += (1/m) * ( k*x_data[j] + b - y_data[j] )

        k = k - lr*k_grad
        b = b - lr*b_grad
    return k,b


Draw_Scatter(X,Y,0,0)
print("------------------------")
k,b = Gradient_Descent(0,0,X,Y,60,0.01)
print('斜率 {0}'.format(k)) 
print(b)

Draw_Scatter(X,Y,k,b)
#Draw_Linear(2,b)
#Least_Square(k,b,X,Y)

    


```
