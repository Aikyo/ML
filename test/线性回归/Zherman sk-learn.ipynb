```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


# data = pd.read_csv("data.csv")
# #print(data)
# X = data["x"]
# Y = data["y"]
# z = data[:,"x"]
# #Z = data["z"]
data = np.genfromtxt("data.csv",delimiter = ",")

X = data[1:,0]
Y = data[1:,1]
#print(X)
print(X.shape)
plt.scatter(X,Y)

# plt.scatter(X,Y)
# plt.show()
```

```python
x_data = data[1:,0,np.newaxis]
y_data = data[1:,1,np.newaxis]
model = LinearRegression()
model.fit(x_data,y_data)

```

```python
plt.plot(X,Y,'r.')
plt.plot(x_data,model.predict(x_data),'b')
print('coefficient is {0}'.format(model.coef_))
```
